Original Matrix: Row Vector: Result Matrix:
[[1 2 3] [10 20 30] [[11 22 33]

9.How can you iterate through each element of a NumPy array using a loop? Provide an
example.

We can iterate through each element of a NumPy array using nested loops, where the outer loop iterates the rows and the inner loop iterates the columns.

```python
               # 0  1  2
arr = np.array([[1, 2, 3], #0
                [4, 5, 6]])#1

for i in range(arr.shape[0]):#for rows
    for j in range(arr.shape[1]):#for columns
        print('Element at ({}, {}): {}'.format(i, j, arr[i, j]))#to print rows and column wise elements
```
